Step 4 : Data Preparation

* Count % of missing value per column

```python
df.isnull().mean()*100
```

* As we understood that there are a lot of null values in PURPOSE column, so for that we will me filling the null values with a NOT keyword.

```python
df['PURPOSE'].fillna("Not", inplace=True)
df.head()
```

* Converting  the START_DATE and END_DATE to the date_time format to do further analysis

```python
df['START_DATE'] = pd.to_datetime(df['START_DATE'], errors = 'coerce' )
df['END_DATE'] = pd.to_datetime(df['END_DATE'], errors='coerce')

df.head()
```

* Feature Engineering: Extracting Time Components

```python
#Extracting Date
df['START_DATE_DAY'] = df['START_DATE'].dt.day
df['START_DATE_time'] = df['START_DATE'].dt.hour

df.head()
```

*Feature Engineering : Convert it into categorical

```python
df['day-night'] = pd.cut(df['START_DATE_time'], bins=[0,10,15,19,24], labels = ['Morning','Afternoon','Evening','Night'])
df.head()
```

* Find duplicate

```python
df.duplicated().sum()
```

* Remove Duplicate and Delete Missing value

```python
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)

df.head()
```
